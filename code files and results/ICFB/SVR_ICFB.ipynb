# Support Vector Regression (SVR)

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('DummyData.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
y = y.reshape(len(y), 1)
```

##Splitting the Dataset

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
# Standardize the features
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train = scaler_X.fit_transform(X_train)
y_train = scaler_y.fit_transform(y_train)
```

```
print(X_train)
```

```
print(y_train)
```

## Training the SVR model on the training dataset

```
from sklearn.svm import SVR

# Creating and training the SVR model
regressor = SVR(kernel='rbf')
regressor.fit(X_train, y_train)
```

##Predicting the Test set results

```
y_pred = scaler_y.inverse_transform(regressor.predict(scaler_X.transform(X_test)).reshape(-1,1))
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Predictions of the results using Trained Dataset

```
new_data = [[470, 0.5, 0.119, 2.32]]  # Values taken from the test set and the actual ouput was 5.7

# Standardize the new data using the previously fitted scaler
new_data_scaled = scaler_X.transform(new_data)

# Make predictions
new_pred_scaled = regressor.predict(new_data_scaled)

# Inverse transform the prediction to the original scale
new_pred = scaler_y.inverse_transform(new_pred_scaled.reshape(-1, 1)).flatten()

print(f'Predicted output: {new_pred}')
```

## Evaluating the model performance on the Testset

```
from sklearn.metrics import mean_squared_error, r2_score
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))
```
